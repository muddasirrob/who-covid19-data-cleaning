**Data Cleaning for a Real-world Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('worldometer_data.csv')
df.head()
```

**Summary Statistics and Data Info**

```python
df.info()
df.describe()
df.isnull().sum()
```

**Visualization: Missing value heatmap**


```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

**Step 3: Handle Missing Values**

```python
df['Population'] = df['Population'].fillna(df['Population'].median())
```

**Step 4: Remove Duplicates**

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

**Step 5: Fix Inconsistencies**

```python
df.columns = df.columns.str.strip()
df['Continent'].unique()
```

**Step 6: Data Type Conversion**

```python
df['Population'] = df['Population'].astype(int)
```

**Final Review**

```python
df.info()
df.head()
```

**Visualization of Cleaned Data**

```python
print(df.columns)
top10 = df.sort_values(by='TotalCases', ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='Country/Region', y='TotalCases', data=top10)
plt.xticks(rotation=45)
plt.title("Top 10 Countries by Total COVID-19 Cases")
plt.show()
df.columns = df.columns.str.strip()
```

```python
print(df.columns)
plt.figure(figsize=(10,6))
sns.boxplot(x='Continent', y='Tot Cases/1M pop', data=df)
plt.title("Distribution of COVID-19 Cases Per Million by Continent")
plt.xticks(rotation=45)
plt.ylabel("Cases Per Million")
plt.show()
```
